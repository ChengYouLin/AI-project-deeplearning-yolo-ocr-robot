```python
results = metrics.class_result  # Êàñ metrics.class_result
for i, res in enumerate(results):
    p, r, ap50, ap = res  # precision, recall, AP@0.5, AP@0.5:0.95
    print(f"Class {i}: P={p:.3f}, R={r:.3f}, AP@0.5={ap50:.3f}, AP@0.5:0.95={ap:.3f}")
```

```python
import numpy as np

iou_thresholds = np.linspace(0.5, 0.95, 10)
for class_idx, class_ap in enumerate(metrics.all_ap):
    print(f"Class {metrics.ap_class_index[class_idx]}:")
    for i, ap in enumerate(class_ap):
        print(f"  IoU {iou_thresholds[i]:.2f}: AP = {ap:.3f}")
```

```python
from ultralytics import YOLO
import easyocr
import cv2
import matplotlib.pyplot as plt

# ËÆÄÂèñÊ®°Âûã
model = YOLO('/content/drive/MyDrive/AIProjectYOLO/runs/yolov11n_plate4/weights/best.pt')

# ËºâÂÖ•Ê∏¨Ë©¶ÂúñÁâáÔºàÊõ¥ÊèõÁÇ∫‰Ω†‰∏äÂÇ≥ÁöÑÂúñÊ™îÂêçÔºâ
img_path = '11baed9f980242ac110005.jpg'  # ‚Üê ÊîπÊàê‰Ω†ÁöÑÂúñÁâáÊ™îÂêç
image = cv2.imread(img_path)

# Êé®Ë´ñ
results = model(image)[0]  # ÂèñÂæóÁ¨¨‰∏ÄÂºµÂúñÁâáÁöÑÁµêÊûú

# Âª∫Á´ã OCR Ê®°ÂûãÔºàÊîØÊè¥‰∏≠Ëã±ÊñáÔºâ
reader = easyocr.Reader(['en', 'ch_sim'])

# ËôïÁêÜÊØè‰∏ÄÂÄãÂÅµÊ∏¨Âà∞ÁöÑËªäÁâåÊ°Ü
for box in results.boxes:
    xyxy = box.xyxy[0].cpu().numpy().astype(int)
    x1, y1, x2, y2 = xyxy
    cropped = image[y1:y2, x1:x2]  # Ë£ÅÂàáËªäÁâåÂçÄÂüü

    # OCR ËÆÄÂèñÊñáÂ≠ó
    ocr_result = reader.readtext(cropped)
    print(f"üîç YOLO ÂÅµÊ∏¨Âà∞ÂçÄÂüüÔºö({x1}, {y1}) - ({x2}, {y2})")
    for (bbox, text, conf) in ocr_result:
        print(f"üöò ËªäÁâåËæ®Ë≠òÁµêÊûúÔºö{text}Ôºà‰ø°ÂøÉÔºö{conf:.2f}Ôºâ")

    # È°ØÁ§∫Ë£ÅÂàáÁµêÊûú
    plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
    plt.title(f"OCR: {text}")
    plt.axis('off')
    plt.show()
```
