```python
metrics = model.val()
print(metrics.all_ap.shape)  # (nc, 10)，表示每個 class 在 10 個 IOU threshold 下的 AP
```

```python
results = metrics.class_result  # 或 metrics.class_result
for i, res in enumerate(results):
    p, r, ap50, ap = res  # precision, recall, AP@0.5, AP@0.5:0.95
    print(f"Class {i}: P={p:.3f}, R={r:.3f}, AP@0.5={ap50:.3f}, AP@0.5:0.95={ap:.3f}")
```

```python
import numpy as np

iou_thresholds = np.linspace(0.5, 0.95, 10)
for class_idx, class_ap in enumerate(metrics.all_ap):
    print(f"Class {metrics.ap_class_index[class_idx]}:")
    for i, ap in enumerate(class_ap):
        print(f"  IoU {iou_thresholds[i]:.2f}: AP = {ap:.3f}")
```

```python
from ultralytics import YOLO
import easyocr
import cv2
import matplotlib.pyplot as plt

# 讀取模型
model = YOLO('/content/drive/MyDrive/AIProjectYOLO/runs/yolov11n_plate4/weights/best.pt')

# 載入測試圖片（更換為你上傳的圖檔名）
img_path = '11baed9f980242ac110005.jpg'  # ← 改成你的圖片檔名
image = cv2.imread(img_path)

# 推論
results = model(image)[0]  # 取得第一張圖片的結果

# 建立 OCR 模型（支援中英文）
reader = easyocr.Reader(['en', 'ch_sim'])

# 處理每一個偵測到的車牌框
for box in results.boxes:
    xyxy = box.xyxy[0].cpu().numpy().astype(int)
    x1, y1, x2, y2 = xyxy
    cropped = image[y1:y2, x1:x2]  # 裁切車牌區域

    # OCR 讀取文字
    ocr_result = reader.readtext(cropped)
    print(f"🔍 YOLO 偵測到區域：({x1}, {y1}) - ({x2}, {y2})")
    for (bbox, text, conf) in ocr_result:
        print(f"🚘 車牌辨識結果：{text}（信心：{conf:.2f}）")

    # 顯示裁切結果
    plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
    plt.title(f"OCR: {text}")
    plt.axis('off')
    plt.show()
```

```python
import os

train_path = '/content/drive/MyDrive/AIProjectYOLO/images/train'
val_path = '/content/drive/MyDrive/AIProjectYOLO/images/val'

train_images = [f for f in os.listdir(train_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
val_images = [f for f in os.listdir(val_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

print(f"訓練集張數: {len(train_images)}")
print(f"驗證集張數: {len(val_images)}")

ratio = len(train_images) / (len(train_images) + len(val_images))
print(f"訓練/驗證比例 = {ratio:.2%} / {1-ratio:.2%}")
```
